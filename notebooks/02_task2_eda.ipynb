## First-day Segmentation

```python
import datetime as dt

cutoff = dt.date(2024, 2, 15)
cohort_df = df[df["install_date"] >= cutoff].copy()

print("Cohort users:", cohort_df["user_id"].nunique())
```

```python
d0_df = cohort_df[cohort_df["days_since_install"] == 0].copy()
print("D0 users:", d0_df["user_id"].nunique())
```

```python
d0_users = (
    d0_df
    .groupby("user_id", as_index=False)
    .agg(
        platform=("platform", "first"),
        country=("country", "first"),
        d0_sessions=("total_session_count", "sum"),
        d0_duration=("total_session_duration", "sum"),
        d0_matches=("match_end_count", "sum"),
        d0_victories=("victory_count", "sum"),
        d0_revenue=("total_revenue", "sum"),
        d0_errors=("server_connection_error", "sum")
    )
)

d0_users.shape
```

```python
d0_users["d0_victory_rate"] = (
    d0_users["d0_victories"] /
    d0_users["d0_matches"].replace(0, np.nan)
).fillna(0)
```

```python
d0_users.describe().T
```

```python

d0_users["d0_matches"].quantile([0.5, 0.75, 0.9])
```

```python
d0_users["d0_sessions"].quantile([0.5, 0.75, 0.9])
```

```python
d0_users["d0_duration"].quantile([0.5, 0.75, 0.9])
```

```python
q_duration = d0_users["d0_duration"].quantile(0.75)

def assign_segment(row):
    if row["d0_revenue"] > 0:
        return "Monetizer"
    elif row["d0_matches"] >= 5 and row["d0_victory_rate"] >= 0.6:
        return "Competitive"
    elif row["d0_duration"] >= q_duration or row["d0_matches"] >= 5:
        return "Highly Engaged"
    elif row["d0_sessions"] == 1 and row["d0_duration"] < 600:
        return "Early Drop-off Risk"
    else:
        return "Casual"

d0_users["segment"] = d0_users.apply(assign_segment, axis=1)
```

```python
d0_users["segment"].value_counts(normalize=True) * 100
```

```python
import numpy as np

d0_users = (
    d0_df
    .groupby("user_id", as_index=False)
    .agg(
        platform=("platform", "first"),
        country=("country", "first"),
        d0_sessions=("total_session_count", "sum"),
        d0_duration=("total_session_duration", "sum"),
        d0_matches=("match_end_count", "sum"),
        d0_victories=("victory_count", "sum"),
        d0_iap=("iap_revenue", "sum"),
        d0_ad=("ad_revenue", "sum"),
        d0_errors=("server_connection_error", "sum"),
    )
)

d0_users["d0_revenue"] = d0_users["d0_iap"] + d0_users["d0_ad"]

d0_users["d0_victory_rate"] = (
    d0_users["d0_victories"] / d0_users["d0_matches"].replace(0, np.nan)
).fillna(0)
```

```python
q_duration = d0_users["d0_duration"].quantile(0.75)

def assign_segment(row):
    if row["d0_iap"] > 0:
        return "IAP Monetizer"
    elif row["d0_matches"] >= 5 and row["d0_victory_rate"] >= 0.6:
        return "Competitive"
    elif row["d0_duration"] >= q_duration or row["d0_matches"] >= 5:
        return "Highly Engaged"
    elif row["d0_sessions"] == 1 and row["d0_duration"] < 600:
        return "Early Drop-off Risk"
    else:
        return "Casual"

d0_users["segment"] = d0_users.apply(assign_segment, axis=1)
```

```python
d0_users["segment"].value_counts(normalize=True) * 100
```

```python
d1_users = cohort_df[cohort_df["days_since_install"] == 1]["user_id"].unique()

d0_users["returned_d1"] = d0_users["user_id"].isin(d1_users)
```

## Revenue contrubiton by Segment

```python
user_total_revenue = (
    cohort_df
    .groupby("user_id", as_index=False)["total_revenue"]
    .sum()
)
```

```python
segment_revenue = d0_users.merge(
    user_total_revenue,
    on="user_id",
    how="left"
)

segment_revenue["total_revenue"] = segment_revenue["total_revenue"].fillna(0)
```

```python
segment_summary = (
    segment_revenue
    .groupby("segment")
    .agg(
        users=("user_id", "count"),
        total_revenue=("total_revenue", "sum"),
        avg_revenue_per_user=("total_revenue", "mean")
    )
    .sort_values("total_revenue", ascending=False)
)

segment_summary["revenue_share_%"] = (
    segment_summary["total_revenue"] /
    segment_summary["total_revenue"].sum()
) * 100

segment_summary
```

```python
user_iap = cohort_df.groupby("user_id")["iap_revenue"].sum().reset_index()
user_ad = cohort_df.groupby("user_id")["ad_revenue"].sum().reset_index()

segment_revenue = segment_revenue.merge(user_iap, on="user_id", how="left")
segment_revenue = segment_revenue.merge(user_ad, on="user_id", how="left")

segment_revenue[["iap_revenue","ad_revenue"]] = \
    segment_revenue[["iap_revenue","ad_revenue"]].fillna(0)

segment_mix = (
    segment_revenue
    .groupby("segment")
    .agg(
        total_iap=("iap_revenue", "sum"),
        total_ad=("ad_revenue", "sum")
    )
)

segment_mix
```

## Revenue Contribution Analysis  Key Insights

### Revenue Distribution by Segment ###

Our cohort-based first-day segmentation reveals the following revenue structure:

Competitive segment contributes the largest share of total revenue (~42%).

IAP Monetizers, while representing less than 1% of users, contribute ~38% of total revenue.

Early Drop-off Risk users account for ~43% of users but only ~7% of revenue.

### Key Observations ###

Competitive Users Drive Sustainable Revenue

The Competitive segment:

Shows strong D1 retention (~54%)

Generates the highest total revenue share (~42%)

This indicates that:

The win experience and competitive engagement are strong drivers of both retention and monetization.

Improving matchmaking fairness, skill progression, and reward systems may further increase revenue.

### IAP Monetizers Represent a Whale Segment ###

Only ~0.4% of users

Extremely high ARPU (~17.8)

Contribute ~38% of total revenue

This reflects a typical Pareto distribution common in free-to-play mobile games.

However, revenue is not solely dependent on whales  engagement-driven monetization (ads + competitive behavior) plays a significant role.

### Early Drop-off Risk Segment Represents a Major Opportunity ###

43% of users fall into this segment

Very low D1 retention (~19%)

Minimal revenue contribution (~7%)

This suggests potential onboarding or early gameplay friction.

Improving tutorial design, early reward loops, or difficulty balance may reduce churn and significantly increase total revenue.

### Casual Segment Optimization Opportunity ###

Casual users represent a meaningful portion of the player base but generate limited revenue.

Optimizing ad frequency, introducing soft conversion offers, or engagement nudges could increase monetization in this group.

### Strategic Product Implications ###

Based on this analysis:

Enhancing competitive engagement could simultaneously increase retention and revenue.

Reducing early drop-off could unlock significant monetization potential.

Monetization strategy should balance whale targeting with engagement-driven revenue growth.

### Core Insight ###

While IAP monetizers generate high revenue per user, the Competitive segment contributes the largest share of total revenue due to its size and strong retention.
Therefore, strengthening the win experience appears to be the most scalable growth lever.

### Session Duration Lifecycle Trend  Key Findings ###

Average and median session duration both increase over time.

After an initial drop between Days 25, session duration shows a steady upward trend.

This suggests that retained users become increasingly engaged as they progress through the game.

The trend is visible in both mean and median values, indicating that the increase is not driven solely by heavy users.

### Interpretation ###

This pattern reflects a healthy lifecycle dynamic:

Early churn removes low-engagement users.

Remaining users develop stronger engagement habits.

Competitive or progression-based mechanics may be reinforcing longer sessions.

## Platform Performance ##

```python
platform_summary = (
    cohort_df
    .groupby("platform")
    .agg(
        users=("user_id", "nunique"),
        avg_session_duration=("total_session_duration", "mean"),
        avg_session_count=("total_session_count", "mean"),
        avg_matches=("match_end_count", "mean"),
        total_iap=("iap_revenue", "sum"),
        total_ad=("ad_revenue", "sum"),
        total_errors=("server_connection_error", "sum"),
        total_match_start=("match_start_count", "sum"),
        total_match_end=("match_end_count", "sum"),
        total_victories=("victory_count", "sum")
    )
)

platform_summary["arpu"] = (
    (platform_summary["total_iap"] + platform_summary["total_ad"]) /
    platform_summary["users"]
)

platform_summary["match_completion_rate"] = (
    platform_summary["total_match_end"] /
    platform_summary["total_match_start"]
)

platform_summary["win_rate"] = (
    platform_summary["total_victories"] /
    platform_summary["total_match_end"]
)

platform_summary["error_per_user"] = (
    platform_summary["total_errors"] /
    platform_summary["users"]
)

platform_summary
```

```python
plt.figure(figsize=(6,4))
plt.bar(platform_summary.index, platform_summary["arpu"])
plt.title("ARPU by Platform")
plt.show()
```

### Platform Performance Analysis  Key Findings ###

Android has a significantly larger user base.

iOS users show higher engagement (longer sessions and more matches).

iOS ARPU is approximately 3x higher than Android.

Technical stability metrics (error rate and match completion) are similar across platforms.

Monetization is significantly stronger on iOS, while Android provides scale advantage.


The data suggests a dual-platform strategy: optimize monetization experiments on iOS, while focusing on retention and engagement scaling on Android to unlock long-term revenue growth.
